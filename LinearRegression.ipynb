```python
def gd(X,y,alpha,theta,it):
    m=np.size(y)
    c1= cost(X,y,theta)
    for i in range(0,it):
        hx = np.matmul(X,theta)
        temp = (float(alpha)/m)*(np.matmul(X.T,(hx-y)))
        c2= cost(X,y,theta-temp)
        if(c2<c1):
            c1=c2
            theta=theta-temp
        print(c1)    
    return theta
```

```python
def cost(X,y,theta):
    m=np.size(y)
    hx = np.matmul(X,theta)
    #F= float((1/(2*m))*(np.matmul((np.subtract(hx,y)).T,np.subtract(hx,y))))
    
    return (float(np.sum(np.power(np.subtract(hx,y),2)))/(2*m))
    
```

```python
x = X_train
m = y_train.size
X_train = np.hstack(((np.matrix(np.ones(m).reshape(m,1))),x))
theta = np.matrix(np.ones(2).reshape(2,1))
```

```python
theta = gd(X_train, y_train,0.001, theta, 1500)
c = cost(X_train,y_train,theta)
print(c)
```

```python
plt.scatter(np.array(x), np.array(y_train), marker = 'x', color='r')
plt.plot(np.array(x),np.array(np.matmul(X_train, theta)))
plt.title("After Regression: slope = " + str(theta))
plt.show()
print(cost(X_train, y_train, theta))
```
